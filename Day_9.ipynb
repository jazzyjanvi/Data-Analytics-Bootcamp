```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# ci class interval
# Sample dataset (like 'tips')
data = {
    'total_bill':[10,20,30,40,50,15,25,35,45,55],
    'tip':[1,2,3,4,5,1.5,2.5,3.5,4.5,5.5],
    "sex":['Male','Female']*5,
    "smoker":['Yes','No']*5,
    "day":['Thur','Fri','Sat','Sun','Thur','Fri','Sat','Sun','Tue','Mon'],
    "time":['Lunch','Dinner']*5,
    "size":[1,2,3,4,1,2,3,4,2,3]
}
df = pd.DataFrame(data)

# Set Style 
sns.set(style="whitegrid")
```

```python
sns.boxplot(x='total_bill', data=df)
plt.title("Boxplot")
plt.show()
```

```python
# 1 Barplot
sns.barplot(x="day",y="total_bill",data=df)
plt.title("BarPlot")
plt.show()
```

```python
# 2. count plot
sns.countplot(x="day",data=df)
plt.title("Count Plot")
plt.show()
```

```python
# 3. Strip Plot
sns.stripplot(x="day",y="total_bill",data=df)
plt.title("Strip Plot")
plt.show()
```

```python
# 4. Violin Plot
sns.violinplot(x="day",y="total_bill",data=df)
plt.title("Violin Plot")
plt.show()
```

```python
# striptplot advance version of sactter plot
# 5. Stripplot
sns.stripplot(x="day",y="total_bill",data=df)
plt.title("Stript plot")
plt.show()
```

```python
# swarmplot length kam ho jayega so space kam hoga 
# 6.Swarmplot
sns.swarmplot(x="day",y="total_bill",data=df)
plt.title("Swarmp plot")
plt.show()
```

```python
# 7 Pointplot
sns.pointplot(x="day",y="total_bill",data=df)
plt.title("Stript plot")
plt.show()
```

```python
# 8. boxenplot
# upgrade version of boxplot
# use to distribution for large dataset
# ratio affect in boxplot
# boxplot data divide 
```

```python
# 8. boxenplot
sns.boxenplot(x="day",y="total_bill",data=df)
plt.title("boxen plot")
plt.show()
```

```python
# KDE PLOT FOR CURVE 
# IN CIRCLE FORMAT
# HYPERBOLE AND PARABOLA 
# 9. boxenplot
sns.kdeplot(x="size",y="total_bill",data=df)
plt.title("KDE plot")
plt.show()
```
