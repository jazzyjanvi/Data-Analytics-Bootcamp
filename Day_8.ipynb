```python
import pandas as pd
# panel data - measured value
# data maunpication
# series and  data forms
```

```python
# from dictionary 
data = {
    "Name":["Janvi","Sarabjeet","Ujjwal"],
    "Age":[18,19,18],
    "Marks":[90,99,100]
}
df = pd.DataFrame(data)
print(df)
```

```python
#df = pd.read_csv("dataset.csv") # read_excel 
```

```python
print("Staring 5 rows:\n",df.head()) #starting 5 line
print("Ending 5 rows:\n",df.tail()) #ending 5 line
print("(rows,columns):\n",df.shape) #(rows,column)
print("Columns name:\n",df.columns) #column name
print("Info:\n",df.info()) #summary
print("Stats Summary:\n",df.describe()) #stats summary
```

```python
print(df["Name"])
```

```python
print(df[["Name","Marks"]])
```

```python
print(df.loc[0])
print(df.loc[1])
```

```python
print(df[df["Marks"]>85])
print(df[(df["Marks"]>90) & (df["Age"]<20)])
```

```python
df['Grade']=['A','B','C']
df['Marks']=df['Marks']+5
print(df)
```

```python
#Columnwise Drop
df.drop('Grade', axis=1, inplace=True,errors='ignore') #inplace=true means original dataframe changed, if not there then everywhere same named column deleted
#Rowwise drop
df.drop(1,axis=0,inplace=True)
print(df.columns)
```

```python
#sorting
print(df.sort_values(by="Marks",ascending=False))
```

```python
df=pd.DataFrame({'Name':['a','b',None],'Marks':[1,2,3]})
print(df)
df.fillna('Unknown',inplace=True)
df.dropna(inplace=True)
print()
print(df)
```

```python
df=pd.DataFrame({'Id':[11,22,33,44,55],'Name':['A','B','C','D','E']})
df=pd.DataFrame({'Id':[11,22,33,44,55],'Marks':[93,94,95,96,97]})
print("Dataframe 1:")
print(df)
print()
print("Dataframe 2:")
print(df)
merged=pd.merge(df,df, on='Id')
print()
print("Mergeed Dataframe:")
print(merged)
```

```python
df=pd.DataFrame({'Dept':['IT','HR','HR','IT'], 'Salary':[10,20,30,40]})
print(df.groupby('Dept')['Salary'].mean())
```

```python
#matploting
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [1,2,9,16]
plt.plot(x,y,color = 'Blue',linestyle='--',marker='o')
plt.title("line plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
#Bar Chart
x = [1,2,3,4]
y = [1,2,9,16]
plt.bar(x,y,color="Violet")
plt.title("Bar Chart")
plt.show()
```

```python
#horizontal Bar graph 
x = [1,2,3,4]
y = [1,2,9,16]
plt.barh(x,y,color="Violet")
plt.title("Bar Chart horizontal")
plt.show()
```

```python
#Scatter chart
x = [1,2,3,4]
y = [1,2,9,16]
plt.scatter(x,y,color="Violet")
plt.title("Sactter plot Chart")
plt.show()
```

```python
#histograpy
import numpy as np
data = np.random.randn(1000)
plt.hist(data,bins=30,color='Green')
plt.show()
```

```python
#Pie chart
labels = ["Python","Java","C++"]
size = [60,70,20]
plt.pie(size,labels=labels,autopct="%1.1f%%",startangle=90)
plt.title("Pie Chart")
plt.show()
```
